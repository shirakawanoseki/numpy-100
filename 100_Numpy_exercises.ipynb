#### 25. Given a 1D array, negate all elements which are between 3 and 8, in place. (★☆☆)

````python
rng = np.random.default_rng()
A = rng.integers(low=0, high=50,size=50)
def f(x):
    if 3 < x and x <8:
        return -x
    else:
        return x
vf = np.vectorize(f)
vf(A)
````
